## **Sargassum Detection project**

This notebook contains the definition of a Convolutional Neural Network used to classify satelital images in percentage of sargassum.

```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd

# For dataset management
import glob
import random
import torch.utils.data as data
from dataset import SatelitalImages
```

```python
"""
Dataset loading...
This cell contains the logics for the dataset loading in-memory
using the torch.utils.data library for performance.
"""
paths: list[str] = list()
labels: list[str] = list()

# Add the images with sargassum
with_sargassum = glob.glob("static/with/*.tif")
paths.extend(with_sargassum)
labels.extend(["Yes" for _ in range(len(with_sargassum))])

# Add the images without sargassum
without = glob.glob("static/without/*.tif")
paths.extend(without)
labels.extend(["No" for _ in range(len(without))])

# Randomly shuffle the data
random.shuffle(paths)
random.shuffle(labels)

# Create the object to use
DATA = SatelitalImages(paths, labels)
```

```python
"""
Convolutional Neural Network definition
"""
class SargassumDetector(nn.Module):
    # Inheritance from the neural network module
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.model = nn.Sequential(
            # Define the convolutional layers of the network
            nn.Conv2d(in_channels=3, out_channels=16, kernel_size=3),
            nn.ReLU(),
            nn.AvgPool2d(kernel_size=3, stride=1),
            nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3),
            nn.ReLU(),
            nn.AvgPool2d(kernel_size=3, stride=1),
            nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3),
            nn.ReLU(),
            nn.AvgPool2d(kernel_size=3, stride=1),
            # Flatten the output of the convolution
            nn.Flatten()
            # Define the fully conected 
        )
        
    def forward(self, x):
        return self.model(x)
```
