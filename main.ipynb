## **Sargassum Detection project**

This notebook contains the definition of a Convolutional Neural Network used to classify satelital images in percentage of sargassum.

```python
import torch
import torch.nn as nn
import rasterio
import numpy as np
import pandas as pd
```

```python
def extract_tensor(path: str) -> torch.tensor:
    """
    Loads a .tiff image and parse it into a tensor
    """
    with rasterio.open(path) as file:
        arr = file.read()
        tensor = torch.tensor(arr)
    return tensor


def normalize_tensor(tensor: torch.tensor) -> torch.tensor:
    """
    Applies the min-max normalization to the dataset
    Avoid weird behavior on the data treatment by making every
    data in the same range of values
    """
    return (tensor - tensor.min()) / (tensor.max() - tensor.min())
```

```python
"""
Neural Network definition
"""
class SargassumDetector(nn.Module):
    # Inheritance from the neural network module
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.model = nn.Sequential(
            nn.Conv3d(3, 3)
        )
        
    def forward(self, x):
        return self.model(x)
```
